## Step 1

### 🔢 Task: Gradient Descent with Squared Loss (One Step)

You are given:

* Input `x = 3`
* True output `y_true = 10`
* Initial weight `w = 0.5`
* Learning rate `η = 0.1`
* Model:

  $$
  y_{\text{pred}} = w \cdot x
  $$
* Loss:

  $$
  L = (y_{\text{pred}} - y_{\text{true}})^2
  $$

---

### ✍️ What to Compute:

1. Compute the predicted value: $y_{\text{pred}}$
2. Compute the error: $y_{\text{pred}} - y_{\text{true}}$
3. Compute the loss: $L$
4. Compute the gradient of loss w\.r.t. weight: $\frac{dL}{dw}$
5. Update the weight using gradient descent

   $$
   w_{\text{new}} = w - \eta \cdot \frac{dL}{dw}
   $$
6. What is the new value of $w$?

## Step 1: Solution

### Given:

* $x = 3$
* $y_{\text{true}} = 10$
* Initial weight $w = 0.5$
* Learning rate $\eta = 0.1$
* Prediction: $y_{\text{pred}} = w \cdot x$
* Loss: $L = (y_{\text{pred}} - y_{\text{true}})^2$

---

### Step 1: Compute prediction

$$
y_{\text{pred}} = w \cdot x = 0.5 \cdot 3 = 1.5
$$

---

### Step 2: Compute error

$$
\text{error} = y_{\text{pred}} - y_{\text{true}} = 1.5 - 10 = -8.5
$$

---

### Step 3: Compute loss

$$
L = (-8.5)^2 = 72.25
$$

---

### Step 4: Compute gradient of loss w\.r.t. weight

We’ll use the chain rule:

$$
\frac{dL}{dw} = \frac{dL}{dy_{\text{pred}}} \cdot \frac{dy_{\text{pred}}}{dw}
$$

1. $\frac{dL}{dy_{\text{pred}}} = 2 (y_{\text{pred}} - y_{\text{true}}) = 2 \cdot (-8.5) = -17$
2. $\frac{dy_{\text{pred}}}{dw} = x = 3$

So:

$$
\frac{dL}{dw} = -17 \cdot 3 = -51
$$

---

### Step 5: Gradient descent update

$$
w_{\text{new}} = w - \eta \cdot \frac{dL}{dw} = 0.5 - 0.1 \cdot (-51) = 0.5 + 5.1 = 5.6
$$

---

### ✅ Final Answer:

$$
\boxed{w = 5.6}
$$

# Chapter 2: Let's plot how weight (W) changes

````python
import numpy as np
import matplotlib.pyplot as plt

# Initialize variables
w = 0.0
x = 3
y_true = 10
lr = 0.1
steps = 30

# Record weight values for plotting
w_history = []

for step in range(steps):
    y_pred = w * x
    error = y_pred - y_true
    grad = 2 * error * x
    w = w - lr * grad
    w_history.append(w)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(range(steps), w_history, marker='o')
plt.axhline(y=10/x, color='r', linestyle='--', label='Target w = {:.2f}'.format(10/x))
plt.title("Convergence of w during Gradient Descent")
plt.xlabel("Iteration")
plt.ylabel("w value")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

## 🧨 What is Exploding Gradient?

When gradients during backpropagation become **too large**, parameters like $w$ can grow exponentially and diverge (go to infinity or NaN). This usually happens in deep networks or with very steep loss surfaces.

We’ll demonstrate this with a **simple example** where the gradient becomes too large.

---

## 🔢 Setup: Exaggerated Case

Let’s define:

* Input:         $x = 100$
* True label:     $y_{\text{true}} = 200$
* Prediction:     $y_{\text{pred}} = w \cdot x$
* Loss:         $L = (y_{\text{pred}} - y_{\text{true}})^2$
* Gradient:     $\frac{dL}{dw} = 2 (w x - y) \cdot x$
* Learning rate:  $\eta = 0.01$

---

## 🎯 Start with initial weight

Let’s choose $w = 1$

Then:

### 🔁 First iteration

1. Prediction:

   $$
   y = w \cdot x = 1 \cdot 100 = 100
   $$
2. Loss:

   $$
   L = (100 - 200)^2 = 10000
   $$
3. Gradient:

   $$
   \frac{dL}{dw} = 2 (100 - 200) \cdot 100 = -20000
   $$
4. Update:

   $$
   w = w - \eta \cdot \frac{dL}{dw} = 1 - 0.01 \cdot (-20000) = 1 + 200 = \boxed{201}
   $$

---

### 🔁 Second iteration

1. Prediction:

   $$
   y = 201 \cdot 100 = 20100
   $$
2. Loss:

   $$
   L = (20100 - 200)^2 \approx 3.92 \times 10^8
   $$
3. Gradient:

   $$
   \frac{dL}{dw} = 2 (20100 - 200) \cdot 100 = 3980000
   $$
4. Update:

   $$
   w = 201 - 0.01 \cdot 3980000 = 201 - 39800 = \boxed{-39599}
   $$

---

### 🔁 Third iteration (even worse!)

1. Prediction:

   $$
   y = -39599 \cdot 100 = -3.9599 \times 10^6
   $$
2. Loss:

   $$
   L = (-3.96 \times 10^6 - 200)^2 \gg 10^{13}
   $$
3. Gradient:

   $$
   \frac{dL}{dw} \sim 2 \cdot (-3.96 \times 10^6 - 200) \cdot 100 \sim -7.92 \times 10^8
   $$

---

## 📉 Result: Gradients explode

Every update becomes:

* Hugely overcorrected
* Worse than before
* Loss skyrockets
* Eventually: `NaN` (not-a-number) or overflow

---

## 🛑 How to prevent exploding gradients?

* **Gradient clipping:** clip gradients within a fixed range
* **Weight initialization:** proper scaling (e.g. Xavier, Kaiming)
* **Smaller learning rates**
* **Normalized inputs**
* **Better architectures** (e.g., residual connections, layer norm)

````python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y_true=10, lr=0.001, steps=100, w_init=0.0):
    """Run gradient descent and return weight history"""
    w = w_init
    w_history = [w]

    for step in range(steps):
        y_pred = w * x
        error = y_pred - y_true
        grad = 2 * error * x
        w = w - lr * grad
        w_history.append(w)

        # Stop if values become too large (divergence detection)
        if abs(w) > 1000:
            break

    return w_history

# First: Show the original comparison (x=9,10,11 with lr=0.001)
x_values = [9, 10, 11]
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for idx, x in enumerate(x_values):
    w_history = gradient_descent(x, lr=0.001, steps=50)
    target_w = 10/x

    ax = axes[idx]
    ax.plot(w_history, marker='o', markersize=4, linewidth=2)
    ax.axhline(y=target_w, color='r', linestyle='--', linewidth=2, label=f'Target w = {target_w:.3f}')
    ax.set_title(f'x = {x}, lr = 0.001', fontsize=12)
    ax.set_xlabel('Iteration')
    ax.set_ylabel('w value')
    ax.grid(True, alpha=0.3)
    ax.legend()

    # Add convergence factor (fixed calculation)
    conv_factor = 1 - 2*0.001*x**2
    ax.text(0.05, 0.95, f'Conv factor: {conv_factor:.3f}',
            transform=ax.transAxes, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

plt.suptitle('Original Comparison: Gradient Descent for x=9,10,11 with lr=0.001', fontsize=14)
plt.tight_layout()
plt.show()

# Second: Show effect of different learning rates with x=10
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
x_test = 10
lr_values = [0.001, 0.005, 0.01, 0.011, 0.015, 0.02]

for idx, lr in enumerate(lr_values):
    row = idx // 3
    col = idx % 3
    ax = axes[row, col]

    w_history = gradient_descent(x_test, lr=lr, steps=50)
    target_w = 10/x_test

    # Determine appropriate y-limits based on the data
    max_w = max(abs(min(w_history)), abs(max(w_history)))
    y_limit = min(max_w * 1.2, 50)  # Cap at reasonable value for visualization

    ax.plot(w_history, marker='o', markersize=3, linewidth=2)
    ax.axhline(y=target_w, color='r', linestyle='--', linewidth=2, label=f'Target w = {target_w:.1f}')
    ax.set_ylim(-y_limit, y_limit)

    # Fixed convergence factor calculation
    conv_factor = 1 - 2*lr*x_test**2
    if abs(conv_factor) < 1:
        stability = "Stable"
        color = 'lightgreen'
    elif abs(conv_factor) == 1:
        stability = "Critical"
        color = 'yellow'
    else:
        stability = "Unstable"
        color = 'lightcoral'

    ax.set_title(f'lr = {lr}, Conv factor = {conv_factor:.3f}', fontsize=11)
    ax.set_xlabel('Iteration')
    ax.set_ylabel('w value')
    ax.grid(True, alpha=0.3)
    ax.legend()

    # Add stability indicator
    ax.text(0.05, 0.05, f'{stability}',
            transform=ax.transAxes, verticalalignment='bottom',
            bbox=dict(boxstyle='round', facecolor=color, alpha=0.8))

plt.suptitle(f'Effect of Learning Rate on Convergence (x = {x_test})', fontsize=14)
plt.tight_layout()
plt.show()

# Third: Show the stability regions
x_range = np.linspace(1, 20, 100)
critical_lr = 1 / (2 * x_range**2)  # Fixed: critical lr is 1/(2x²), not 1/x²

plt.figure(figsize=(12, 8))
plt.plot(x_range, critical_lr, 'b-', linewidth=3, label='Critical lr = 1/(2x²)')
plt.axhline(y=0.001, color='r', linestyle='--', linewidth=2, label='Your lr = 0.001')
plt.fill_between(x_range, 0, critical_lr, alpha=0.3, color='green', label='Stable region (lr < 1/(2x²))')
plt.fill_between(x_range, critical_lr, 0.02, alpha=0.3, color='red', label='Unstable region (lr > 1/(2x²))')

plt.xlabel('x value', fontsize=12)
plt.ylabel('Learning rate', fontsize=12)
plt.title('Stability Analysis: Critical Learning Rate vs Input Value', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.xlim(1, 20)
plt.ylim(0, 0.02)

# Mark the specific x values with correct critical learning rates
for x in [9, 10, 11]:
    critical = 1/(2*x**2)
    plt.plot(x, critical, 'ko', markersize=10)
    plt.text(x, critical + 0.001, f'x={x}\nlr<{critical:.4f}',
             ha='center', va='bottom', fontsize=10,
             bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.show()

# Fourth: Demonstrate the mathematical relationship
print("Mathematical Analysis:")
print("=" * 50)
print("Update rule: w_new = w - lr * 2 * (w*x - y_true) * x")
print("Rearranged: w_new = w(1 - 2*lr*x²) + 2*lr*x*y_true")
print()
print("For convergence: |1 - 2*lr*x²| < 1")
print("This gives us: lr < 1/(2*x²)")
print()

for x in [9, 10, 11]:
    critical_lr = 1/(2*x**2)
    current_lr = 0.001
    conv_factor = 1 - 2*current_lr*x**2
    print(f"x = {x:2d}: Critical lr < {critical_lr:.5f}, Current lr = {current_lr:.3f}")
    print(f"        Convergence factor = {conv_factor:.3f} ({'Stable' if abs(conv_factor) < 1 else 'Unstable'})")
    print()
````

Let me explain what's happening in this gradient descent analysis step by step.

### The Problem Setup

We're trying to learn a single weight `w` in the equation `y = w * x` to predict a target value `y_true = 10`. The loss function is mean squared error: `loss = (w*x - y_true)²`.

### The Mathematics Behind Convergence

The gradient descent update rule is:
```
w_new = w - lr * gradient
gradient = 2 * (w*x - y_true) * x
```

This can be rearranged into a linear recurrence relation:
```
w_new = w(1 - 2*lr*x²) + 2*lr*x*y_true
```

This is the key insight! The term `(1 - 2*lr*x²)` acts as a **contraction factor** that determines whether the algorithm converges or diverges.

### Stability Analysis

For the algorithm to converge, we need the contraction factor to have absolute value less than 1:
```
|1 - 2*lr*x²| < 1
```

This gives us the critical condition: **lr < 1/(2x²)**

Let's see what this means for different scenarios:

#### Case 1: Stable Convergence (lr = 0.001, x = 9,10,11)

- **x = 9**: Convergence factor = 1 - 2(0.001)(81) = 0.838
- **x = 10**: Convergence factor = 1 - 2(0.001)(100) = 0.800  
- **x = 11**: Convergence factor = 1 - 2(0.001)(121) = 0.758

All factors are between -1 and 1, so all converge! Notice that x=11 has the smallest factor (closest to 0), so it actually converges **fastest**.

#### Case 2: What Happens with Different Learning Rates (x = 10)

- **lr = 0.001**: Factor = 0.800 → Smooth exponential convergence
- **lr = 0.005**: Factor = 0.000 → Converges in exactly one step!
- **lr = 0.01**: Factor = -1.000 → Critical case, oscillates between two values
- **lr = 0.011**: Factor = -1.200 → Unstable oscillation with growing amplitude
- **lr = 0.015**: Factor = -2.000 → Strong divergent oscillation
- **lr = 0.02**: Factor = -3.000 → Explosive divergence

### Why This Behavior Occurs

1. **Positive convergence factor (0 < factor < 1)**: The weight approaches the target from one side in a smooth exponential decay.

2. **Zero convergence factor**: The algorithm finds the exact solution in one step because the update perfectly cancels the current error.

3. **Negative convergence factor (-1 < factor < 0)**: The weight oscillates around the target but still converges, with the oscillation amplitude decreasing each iteration.

4. **Factor = -1**: Perfect oscillation - the weight bounces between two values forever.

5. **Factor < -1 or factor > 1**: The oscillations or deviations grow larger each iteration, leading to divergence.

### The Geometric Interpretation

Think of the loss function as a parabola. The gradient descent is trying to slide down to the bottom:

- **Small learning rate**: Takes tiny steps, always moves toward the minimum
- **Optimal learning rate**: Takes the perfect step size to reach the minimum quickly
- **Too large learning rate**: Overshoots the minimum and bounces back and forth with increasing magnitude

### Practical Implications

1. **The "sweet spot"**: For each input value x, there's an optimal learning rate around `lr = 1/(4x²)` that gives fastest convergence.

2. **Input magnitude matters**: Larger input values require smaller learning rates. This is why feature scaling is important in machine learning!

3. **The trade-off**: There's always a tension between convergence speed and stability. You want the largest learning rate that still guarantees convergence.

### Why x=11 Converges Fastest

This might seem counterintuitive, but with lr=0.001:
- x=11 has convergence factor 0.758 (closest to 0)
- x=10 has convergence factor 0.800  
- x=9 has convergence factor 0.838 (farthest from 0)

The closer the convergence factor is to 0, the faster each iteration reduces the error. It's like having a stronger "pull" toward the correct answer.

This analysis reveals the fundamental relationship between learning rate, input scale, and convergence behavior in gradient descent - principles that apply to much more complex neural networks as well!

# Vectors

* $\vec{x} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$
* $\vec{w} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$
* $y_{\text{true}} = 10$
* Learning rate $\eta = 0.1$

We’re using a **single linear neuron without activation**, predicting:

$$
y_{\text{pred}} = \vec{w}^\top \vec{x}
$$

Loss:

$$
L = (y_{\text{pred}} - y_{\text{true}})^2
$$

---

### ✅ Step 1: Forward pass

$$
y_{\text{pred}} = \vec{w}^\top \vec{x} =
\begin{bmatrix} 1 & 1 \end{bmatrix}
\begin{bmatrix} 1 \\ 2 \end{bmatrix}
= 1 \cdot 1 + 1 \cdot 2 = 3
$$

---

### ✅ Step 2: Compute error

$$
\text{error} = y_{\text{pred}} - y_{\text{true}} = 3 - 10 = -7
$$

---

### ✅ Step 3: Compute gradient of loss w\.r.t. weights

The loss function:

$$
L = (y_{\text{pred}} - y_{\text{true}})^2
= (\vec{w}^\top \vec{x} - y_{\text{true}})^2
$$

Use the chain rule:

$$
\frac{dL}{d\vec{w}} = 2 \cdot \text{error} \cdot \vec{x}
= 2 \cdot (-7) \cdot \begin{bmatrix} 1 \\ 2 \end{bmatrix}
= \begin{bmatrix} -14 \\ -28 \end{bmatrix}
$$

---

### ✅ Step 4: Weight update using gradient descent

$$
\vec{w}_{\text{new}} = \vec{w} - \eta \cdot \frac{dL}{d\vec{w}}
= \begin{bmatrix} 1 \\ 1 \end{bmatrix}
- 0.1 \cdot \begin{bmatrix} -14 \\ -28 \end{bmatrix}
= \begin{bmatrix} 1 + 1.4 \\ 1 + 2.8 \end{bmatrix}
= \begin{bmatrix} 2.4 \\ 3.8 \end{bmatrix}
$$

---

### ✅ Final Answer: Updated weights

$$
\vec{w}_{\text{new}} = \boxed{\begin{bmatrix} 2.4 \\ 3.8 \end{bmatrix}}
$$

---

### 🧪 Optional: Sanity check (new prediction)

$$
y_{\text{pred,new}} = \vec{w}_{\text{new}}^\top \vec{x}
= 2.4 \cdot 1 + 3.8 \cdot 2 = 2.4 + 7.6 = 10
$$

## Step 1: Forward Pass — Multiply Two 2x2 Matrices

Let

$$
A = \begin{bmatrix}1 & 2 \\ 3 & 4 \end{bmatrix}, \quad
B = \begin{bmatrix}5 & 6 \\ 7 & 8 \end{bmatrix}
$$

The product is

$$
C = A \times B = \begin{bmatrix} c_{11} & c_{12} \\ c_{21} & c_{22} \end{bmatrix}
$$

Calculate each element of $C$:

* $c_{11} = 1 \times 5 + 2 \times 7 = 5 + 14 = 19$
* $c_{12} = 1 \times 6 + 2 \times 8 = 6 + 16 = 22$
* $c_{21} = 3 \times 5 + 4 \times 7 = 15 + 28 = 43$
* $c_{22} = 3 \times 6 + 4 \times 8 = 18 + 32 = 50$

So,

$$
C = \begin{bmatrix}19 & 22 \\ 43 & 50 \end{bmatrix}
$$

---

## Step 2: Define a Loss Function

Let's say we have a simple scalar loss function — the sum of all elements in $C$:

$$
L = \sum_{i,j} c_{ij} = 19 + 22 + 43 + 50 = 134
$$

---

## Step 3: Backprop — Compute $\frac{\partial L}{\partial A}$ and $\frac{\partial L}{\partial B}$

### Step 3.1: Compute $\frac{\partial L}{\partial C}$

Since

$$
L = c_{11} + c_{12} + c_{21} + c_{22}
$$

Each $c_{ij}$ contributes equally with derivative 1:

$$
\frac{\partial L}{\partial c_{ij}} = 1
$$

---

### Step 3.2: Compute $\frac{\partial L}{\partial A}$

Recall matrix multiplication rule:

$$
c_{ij} = \sum_k a_{ik} b_{kj}
$$

Chain rule:

$$
\frac{\partial L}{\partial a_{ik}} = \sum_j \frac{\partial L}{\partial c_{ij}} \frac{\partial c_{ij}}{\partial a_{ik}} = \sum_j 1 \times b_{kj} = \sum_j b_{kj}
$$

Let's compute element-wise for $A$:

* For $a_{11}$: $i=1, k=1$

$$
\frac{\partial L}{\partial a_{11}} = b_{11} + b_{12} = 5 + 6 = 11
$$

* For $a_{12}$: $i=1, k=2$

$$
\frac{\partial L}{\partial a_{12}} = b_{21} + b_{22} = 7 + 8 = 15
$$

* For $a_{21}$: $i=2, k=1$

$$
\frac{\partial L}{\partial a_{21}} = b_{11} + b_{12} = 5 + 6 = 11
$$

* For $a_{22}$: $i=2, k=2$

$$
\frac{\partial L}{\partial a_{22}} = b_{21} + b_{22} = 7 + 8 = 15
$$

So,

$$

$$

### Step 3.3: Compute $\frac{\partial L}{\partial B}$

Similarly,

$$
\frac{\partial L}{\partial b_{kj}} = \sum_i \frac{\partial L}{\partial c_{ij}} \frac{\partial c_{ij}}{\partial b_{kj}} = \sum_i 1 \times a_{ik} = \sum_i a_{ik}
$$

Compute element-wise for $B$:

* For $b_{11}$: $k=1, j=1$

$$
\frac{\partial L}{\partial b_{11}} = a_{11} + a_{21} = 1 + 3 = 4
$$

* For $b_{12}$: $k=1, j=2$

* For $b_{21}$: $k=2, j=1$

* For $b_{22}$: $k=2, j=2$
